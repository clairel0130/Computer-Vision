1.2.1 Sharpen using box filter and Gaussian filter

```python
# Xijun(Claire) LIU 260654285
import matplotlib.pyplot as plt
import numpy as np
import cv2

# image filtering
image_path = "./rice.png"
img = cv2.imread(image_path)

# create a box filter
kernel = np.ones((5,5), np.float32)/(5*5)
# box filter and sharpen the image 
box_Blur = cv2.filter2D(img, -1, kernel)
detail_box = img - box_Blur
box_sharpen = img + detail_box

#Gaussian filter and sharpen the image
gaussian_Blur = cv2.GaussianBlur(img,(5,5), np.sqrt(2))
detail_gaussian = img - gaussian_Blur
gaussian_sharpen = img + detail_gaussian


# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
box_sharpen = cv2.cvtColor(box_sharpen, cv2.COLOR_BGR2RGB)
gaussian_sharpen = cv2.cvtColor(gaussian_sharpen, cv2.COLOR_BGR2RGB)

# plot
plt.figure(figsize= (20, 20))
plt.subplot(131), plt.imshow(img)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(box_sharpen)
plt.title("Box Sharpen"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(gaussian_sharpen)
plt.title("Gaussian Sharpen"), plt.xticks([]), plt.yticks([])
plt.show()
```

1.2.2 The box filtered image will have a better sharpening result. As the box filter uses the average value in the "box" as the pixel value of the blurred image, whereas the gaussian filter uses the weighted average value. The weights used in the gaussian filter follows the normal distribution, which indicates that the farther the pixel away from the center, the less weight it get. Thus, the difference between the box filterred image and the original image will be larger than that of the Gaussian filterred image. This means that the 'detail' of the box filterred value will be greater than that of the gaussian filterred one. Thus, the box filterred image will have a better sharpening result. 
In the resulted image, it can be observed that the box filterred image has more details than the Gaussian one, which is consistent with the theoretical result.
