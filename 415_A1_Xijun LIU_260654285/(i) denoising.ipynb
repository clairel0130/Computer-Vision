1.1.1 Gaussian Filtering

```python
# Xijun(Claire) LIU 260654285
import matplotlib.pyplot as plt
import numpy as np
import cv2`

# image filtering
image_path = "./gaussian_noise_peppers.png"
img = cv2.imread(image_path)

# filter image using 7x7 Gaussian, the std is sqrt(2), and the variance is 2
res1 = cv2.GaussianBlur(img,(7,7), np.sqrt(2))

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)

# plot
plt.figure(figsize= (20, 20))
plt.subplot(131), plt.imshow(img)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(res1)
plt.title("7x7 Gaussian"), plt.xticks([]), plt.yticks([])
plt.show()
```

1.1.2 Box Filtering

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# image filtering
image_path = "./gaussian_noise_peppers.png"
img = cv2.imread(image_path)

# create a box filter
kernel = np.ones((7,7), np.float32)/(7*7)
print("kernel:", kernel)

# filter the image
res1 = cv2.filter2D(img, -1, kernel)

# similar results using cv2.blur()
res2 = cv2.blur(img, (7,7))

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)
res2 = cv2.cvtColor(res2, cv2.COLOR_BGR2RGB)

# plot
plt.figure(figsize= (20, 20))
plt.subplot(131), plt.imshow(img)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(res1)
plt.title("Box filter"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(res2)
plt.title("Blur filter"), plt.xticks([]), plt.yticks([])
plt.show()
```

## 1.1.3 PSNR Results

```python
import skimage
from skimage import measure 

imageO_path = "./peppers.png"
image_path = "./gaussian_noise_peppers.png"
imgO = cv2.imread(imageO_path)
img = cv2.imread(image_path)

gaussianFiltered = cv2.GaussianBlur(img,(7,7), np.sqrt(2))
imgO = cv2.cvtColor(imgO, cv2.COLOR_BGR2RGB)
gaussianFiltered = cv2.cvtColor(gaussianFiltered, cv2.COLOR_BGR2RGB)

kernel = np.ones((7,7), np.float32)/(7*7)
boxFiltered = cv2.filter2D(img, -1, kernel)
boxFiltered = cv2.cvtColor(boxFiltered, cv2.COLOR_BGR2RGB)

PSNR_gaussianF = skimage.measure.compare_psnr(imgO, gaussianFiltered)
PSNR_boxF = skimage.measure.compare_psnr(imgO, boxFiltered)

print("PSNR_gaussianF: ", PSNR_gaussianF)
print("PSNR_boxF: ", PSNR_boxF)
```

Statement: the Gaussian filter gives a better denoising result to the Gaussian noise, as higher PSNR indicates a clearer image, where PSNR is the Peak Signal to Noise Ratio.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# image filtering
image_path = "./saltandpepper_noise_peppers.png"
img = cv2.imread(image_path)

# filter image using 7x7 median filter
median_filtered = cv2.medianBlur(img, 7)

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
median_filtered = cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB)

# plot
plt.figure(figsize= (20, 20))
plt.subplot(121), plt.imshow(img)
plt.title("Image"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(median_filtered)
plt.title("Median"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import skimage
from skimage import measure 

imageO_path = "./peppers.png"
image_path = "./saltandpepper_noise_peppers.png"
imgO = cv2.imread(imageO_path)
img = cv2.imread(image_path)

gaussianFiltered = cv2.GaussianBlur(img,(7,7), np.sqrt(2))
imgO = cv2.cvtColor(imgO, cv2.COLOR_BGR2RGB)
gaussianFiltered = cv2.cvtColor(gaussianFiltered, cv2.COLOR_BGR2RGB)

medianFiltered = cv2.medianBlur(img, 7)
medianFiltered = cv2.cvtColor(medianFiltered, cv2.COLOR_BGR2RGB)

PSNR_gaussianF = skimage.measure.compare_psnr(imgO, gaussianFiltered)
PSNR_medianF = skimage.measure.compare_psnr(imgO, medianFiltered)

print("PSNR_gaussianF: ", PSNR_gaussianF)
print("PSNR_MedianF: ", PSNR_medianF)
```

Statement: the Median filter gives a better denoising result to the salt and pepper noise, as higher PSNR indicates a clearer image, where PSNR is the Peak Signal to Noise Ratio.
