3.1 Template Matching with Template1

```python
# Xijun(Claire) LIU 260654285
import matplotlib.pyplot as plt
import numpy as np
import cv2

# load image
img = cv2.imread("where_is_waldo.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# create copy
img_disp = img.copy()

# Read the template
temp = cv2.imread("template1.png")
temp = cv2.cvtColor(temp, cv2.COLOR_BGR2RGB)
# read the size of the template which will be used after found the location of the minimum point
ht, wt = temp.shape[0:2]

# apply template matching
# cv2.matchTemplate(image, templ, method) → result
res = cv2.matchTemplate(img, temp, cv2.TM_SQDIFF_NORMED)

# Display all the SSD in the image
print("SSD: ", res)

#->minVal, maxVal, minLoc, maxLoc
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(res)
# print the location of the minimum SSD point
print("minimum SSD location: ", minLoc)

# circle the template
cv2.rectangle(img_disp, minLoc, (minLoc[0]+wt, minLoc[1]+ht), (255,255,0), 5)

# display images
plt.figure(figsize=(50, 50))
plt.subplot(131), plt.imshow(img)
plt.title("Input Image"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(temp)
plt.title("Template Image"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(img_disp)
plt.title("Matched Image"), plt.xticks([]), plt.yticks([])
plt.show()
```
