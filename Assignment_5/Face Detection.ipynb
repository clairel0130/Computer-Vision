# 2. Eigenvector Representstion

### Use PCA to find the eigenvector representsation of your training data

```python
pca = PCA(n_components=30)
pca.fit(training_img)
print('eigenvectors:')
print(pca.components_)
print('eigenvalues:')
print(pca.explained_variance_)
```

### Plot the variance explained by every vector in descending order (total variance explained vs. number of principal components)

```python
total_variance_explained = []
total= 0
for i in range(30):
    total = total + pca.explained_variance_[i]
    total_variance_explained.append(total)
    
plt.figure(figsize=(10,5))
x = np.arange(30)
plt.plot(x,total_variance_explained)
plt.title("Total Variance Explained vs. Number of Principal Components",fontsize=12)
plt.xlabel("Number of Principal Components",fontsize=10)
plt.ylabel("Variance Explained",fontsize=10)
```

### Do you need all the vectors to represent the data?

No, we don't need all the vectors to represent the data. The larger variance_explain represents more variance of the data and correspond larger eigenvalue, and also smaller recondtruction error.  Thus, the main purpose of choosing the vectors is to find the vectors that contributes more to the variance. In my case, the last eigenvector is extremely small, so it is not important and can be ignored. Also from the plot, it can be observed that the the total-variance-explain curve is convergence to some point. Thus, the only points that have to be considered are values before the convergence. 

### For every sample, find the nearest neighbour (L2 distance), and check whether it is of the same class. What fraction of your data has a neighbour of the same class?

```python
dist = DistanceMetric.get_metric('euclidean')
distance = dist.pairwise(training_img)
min_I = []
min_D = []

for i in range(30):
    minDis = np.inf
    minIndex = -1
    for j in range(30):
        if(distance[i][j] < minDis) and (i != j):
            minDis = distance[i][j]
            minIndex = j 
    min_I.append(minIndex)
    min_D.append(minDis)
print(min_I)

same = 0
for x in range (30):
    if x//10 == min_I[x]//10 :
        same = same + 1
print('In pixel space, fraction of same class is ', same/30)
```

# 3. Face Detection & Identification

### You will now detect the faces in the group image. Use a sliding window to detect whether a section of the image contains a face. Set a threshold on the distance in eigenspace between the window contents and your training data.

```python
group = cv2.imread("Harry.png",  cv2.COLOR_BGR2RGB)
group_gray = cv2.imread("Harry.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(cv2.cvtColor(group, cv2.COLOR_BGR2RGB))
plt.show()
h = 150
w = 150
step = 30
threshold = 13590
windows = []
windows_pos = []

pca = PCA(n_components=30)
pca.fit(training_img)
eigVec = pca.components_
avg = pca.mean_

# Slide through the image
for y in range (0, group.shape[0]-h, step):
    for x in range (0, group.shape[1]-w, step):
        patch = group_gray[y:y+h,x:x+w]
        patch = patch.flatten()
        windows.append(patch)
        windows_pos.append([y, x])

#for all the patches (windows), project them onto the eigenface
eigenface = []
for i in range (len(windows)):
    diff = windows[i] - avg
    summ = avg
    for j in range (eigVec.shape[0]):
        scale = np.dot(diff, eigVec[j])
        summ = summ + scale * eigVec[j]
    eigenface.append(summ)

face_Pos = []
faces = []
for m in range (len(eigenface)):
    dist = DistanceMetric.get_metric('euclidean')
    distance = dist.pairwise([eigenface[m]], eigVec)
    minDis = distance.max()
    minIn = -1
    for n in range(30):
        if(distance[0][n] < minDis):
            minDis = distance[0][n]
            minIn = n 
    if minDis < threshold:
        face_Pos.append(windows_pos[m])
        faces.append(distance)
for i in range(len(face_Pos)):
    [y,x] = face_Pos[i]
    cv2.rectangle(group, (x, y), (x+h, y+w), (0, 255, 0), 2)
    

plt.figure(figsize=(15, 15))
plt.imshow(cv2.cvtColor(group, cv2.COLOR_BGR2RGB)), 
plt.title('Group of three')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# faces is an array used to store all the distances from the detected face to each of the eigenvectors(30 in total)
for i in range (len(faces)):
    minDis = np.inf
    minIn = -1
    for n in range(30):
        if(faces[i][0][n] < minDis):
            minDis = faces[i][0][n]
            minIn = n
    if minIn//10 == 0:
        print('The face at position', face_Pos[i], 'is Harry.')
    if minIn//10 == 1:
        print('The face at position', face_Pos[i], 'is Hermione.')
    if minIn//10 == 2:
        print('The face at position', face_Pos[i], 'is Ron.')
```

# 4. Viola-Jones Comparison

### Use an existing implementation of the Viola-Jones face detector, and compare the results with your detector

```python

# Get OpenCV datapath
data_path = cv2.data.haarcascades
face_cascade = cv2.CascadeClassifier( os.path.join(data_path, 'haarcascade_frontalface_default.xml'))
eye_cascade = cv2.CascadeClassifier( os.path.join(data_path, 'haarcascade_eye.xml'))

img = cv2.imread('Harry.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# detect faces
# detectMultiScape(image, scaleFactor, minNeighbors)
# scaleFactor - parameter specifying how much the image size is reduced
#  at each image scale
# minNeighbors - parameter specifying how many neighbors each candidate
#  rectangle should have to retain
faces = face_cascade.detectMultiScale(img_gray, 1.1,2)
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)
    roi_gray = img_gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 2)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (0,255,0), 2)

# display images
plt.figure(figsize=(15,15))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Group Picture"), plt.xticks([]), plt.yticks([])
plt.show()
```
